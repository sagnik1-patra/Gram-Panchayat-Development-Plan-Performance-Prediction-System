```python
import os
import json
import yaml
import joblib
import warnings
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder, StandardScaler, LabelEncoder
from sklearn.compose import ColumnTransformer
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC

import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.callbacks import EarlyStopping

warnings.filterwarnings("ignore")

# ============================================================
# 1. Paths
# ============================================================

DATASET_PATH = r"C:\Users\NXTWAVE\Downloads\Gram Panchayat Development Plan Performance Prediction System\37ca30d3-ea8c-48a5-a5dc-50ec5f46edd5.csv"

OUTPUT_DIR = r"C:\Users\NXTWAVE\Downloads\Gram Panchayat Development Plan Performance Prediction System"

os.makedirs(OUTPUT_DIR, exist_ok=True)

PREFIX = "hsa_"

H5_MODEL_PATH = os.path.join(OUTPUT_DIR, PREFIX + "model.h5")
PKL_MODEL_PATH = os.path.join(OUTPUT_DIR, PREFIX + "model.pkl")
PREPROCESSOR_PATH = os.path.join(OUTPUT_DIR, PREFIX + "preprocessor.pkl")
LABEL_ENCODER_PATH = os.path.join(OUTPUT_DIR, PREFIX + "label_encoder.pkl")
FEATURE_MASK_PATH = os.path.join(OUTPUT_DIR, PREFIX + "feature_mask.pkl")

YAML_PATH = os.path.join(OUTPUT_DIR, PREFIX + "config.yaml")
JSON_PATH = os.path.join(OUTPUT_DIR, PREFIX + "results.json")

RESULT_CSV_PATH = os.path.join(OUTPUT_DIR, PREFIX + "result.csv")
PREDICTION_CSV_PATH = os.path.join(OUTPUT_DIR, PREFIX + "prediction.csv")

ACCURACY_GRAPH_PATH = os.path.join(OUTPUT_DIR, PREFIX + "accuracy_graph.png")
HEATMAP_PATH = os.path.join(OUTPUT_DIR, PREFIX + "heatmap.png")
COMPARISON_GRAPH_PATH = os.path.join(OUTPUT_DIR, PREFIX + "comparison_graph.png")
RESULT_GRAPH_PATH = os.path.join(OUTPUT_DIR, PREFIX + "result_graph.png")
PREDICTION_GRAPH_PATH = os.path.join(OUTPUT_DIR, PREFIX + "prediction_graph.png")
OPTIMIZATION_GRAPH_PATH = os.path.join(OUTPUT_DIR, PREFIX + "optimization_graph.png")

# ============================================================
# 2. Load Dataset
# ============================================================

print("Loading dataset...")

df = pd.read_csv(DATASET_PATH)
df.columns = df.columns.str.strip()

print("Dataset loaded successfully.")
print("Shape:", df.shape)
print("Columns:", list(df.columns))

# ============================================================
# 3. Validate Required Columns
# ============================================================

required_columns = [
    "stateCode",
    "stateName",
    "districtCode",
    "districtName",
    "finYear",
    "totalGP",
    "sabhaHeld",
    "gpdpApproved",
    "gpdpUpload",
    "date updated"
]

missing_columns = [col for col in required_columns if col not in df.columns]

if missing_columns:
    raise ValueError(f"Missing columns in dataset: {missing_columns}")

# ============================================================
# 4. Data Cleaning
# ============================================================

df["date updated"] = pd.to_datetime(df["date updated"], errors="coerce")

if df["date updated"].isnull().sum() > 0:
    if len(df["date updated"].mode()) > 0:
        df["date updated"] = df["date updated"].fillna(df["date updated"].mode()[0])
    else:
        df["date updated"] = df["date updated"].fillna(pd.Timestamp("2024-01-01"))

df["updated_year"] = df["date updated"].dt.year
df["updated_month"] = df["date updated"].dt.month
df["updated_day"] = df["date updated"].dt.day

numeric_cols = [
    "stateCode",
    "districtCode",
    "totalGP",
    "sabhaHeld",
    "gpdpApproved",
    "gpdpUpload"
]

for col in numeric_cols:
    df[col] = pd.to_numeric(df[col], errors="coerce")
    df[col] = df[col].fillna(df[col].median())

categorical_cols = [
    "stateName",
    "districtName",
    "finYear"
]

for col in categorical_cols:
    df[col] = df[col].astype(str)
    df[col] = df[col].fillna("Unknown")

df["totalGP"] = df["totalGP"].replace(0, np.nan)
df["totalGP"] = df["totalGP"].fillna(df["totalGP"].median())

# ============================================================
# 5. Feature Engineering
# ============================================================

df["sabha_ratio"] = df["sabhaHeld"] / df["totalGP"]
df["approval_ratio"] = df["gpdpApproved"] / df["totalGP"]
df["upload_ratio"] = df["gpdpUpload"] / df["totalGP"]

df["sabha_ratio"] = df["sabha_ratio"].replace([np.inf, -np.inf], 0).fillna(0)
df["approval_ratio"] = df["approval_ratio"].replace([np.inf, -np.inf], 0).fillna(0)
df["upload_ratio"] = df["upload_ratio"].replace([np.inf, -np.inf], 0).fillna(0)

df["performance_score"] = (
    df["sabha_ratio"] +
    df["approval_ratio"] +
    df["upload_ratio"]
) / 3

df["performance_score"] = df["performance_score"].clip(0, 1)

def classify_performance(score):
    if score < 0.50:
        return "Low"
    elif score < 0.80:
        return "Medium"
    else:
        return "High"

df["performance_category"] = df["performance_score"].apply(classify_performance)

print("\nTarget Distribution:")
print(df["performance_category"].value_counts())

# ============================================================
# 6. Features and Target
# ============================================================

feature_columns = [
    "stateCode",
    "stateName",
    "districtCode",
    "districtName",
    "finYear",
    "totalGP",
    "sabhaHeld",
    "gpdpApproved",
    "gpdpUpload",
    "updated_year",
    "updated_month",
    "updated_day",
    "sabha_ratio",
    "approval_ratio",
    "upload_ratio"
]

target_column = "performance_category"

X = df[feature_columns]
y = df[target_column]

label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)

categorical_features = [
    "stateName",
    "districtName",
    "finYear"
]

numeric_features = [
    "stateCode",
    "districtCode",
    "totalGP",
    "sabhaHeld",
    "gpdpApproved",
    "gpdpUpload",
    "updated_year",
    "updated_month",
    "updated_day",
    "sabha_ratio",
    "approval_ratio",
    "upload_ratio"
]

# ============================================================
# 7. Preprocessing
# ============================================================

preprocessor = ColumnTransformer(
    transformers=[
        ("num", StandardScaler(), numeric_features),
        ("cat", OneHotEncoder(handle_unknown="ignore"), categorical_features)
    ]
)

# ============================================================
# 8. Train-Test Split
# ============================================================

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y_encoded,
    test_size=0.2,
    random_state=42,
    stratify=y_encoded
)

X_train_processed = preprocessor.fit_transform(X_train)
X_test_processed = preprocessor.transform(X_test)

X_train_dense = X_train_processed.toarray() if hasattr(X_train_processed, "toarray") else X_train_processed
X_test_dense = X_test_processed.toarray() if hasattr(X_test_processed, "toarray") else X_test_processed

print("\nProcessed Train Shape:", X_train_dense.shape)
print("Processed Test Shape:", X_test_dense.shape)

# ============================================================
# 9. HSA Feature Selection
# ============================================================

print("\nRunning HSA Feature Selection...")

def ensure_non_empty_mask(mask):
    """
    Ensures that at least one feature is selected.
    """
    if np.sum(mask) == 0:
        mask[np.random.randint(0, len(mask))] = 1
    return mask


def evaluate_harmony(mask, X_train_data, X_test_data, y_train_data, y_test_data):
    """
    Fitness function for Harmony Search Algorithm.
    Each harmony is a binary feature-selection mask.
    """

    mask = ensure_non_empty_mask(mask.copy())

    selected_train = X_train_data[:, mask == 1]
    selected_test = X_test_data[:, mask == 1]

    model = RandomForestClassifier(
        n_estimators=80,
        random_state=42,
        class_weight="balanced",
        n_jobs=-1
    )

    model.fit(selected_train, y_train_data)
    predictions = model.predict(selected_test)

    accuracy = accuracy_score(y_test_data, predictions)
    feature_penalty = np.sum(mask) / len(mask)

    fitness = accuracy - (0.01 * feature_penalty)

    return fitness


def hsa_feature_selection(
    X_train_data,
    X_test_data,
    y_train_data,
    y_test_data,
    harmony_memory_size=20,
    iterations=20,
    harmony_memory_considering_rate=0.90,
    pitch_adjusting_rate=0.30,
    random_selection_rate=0.10
):
    """
    Harmony Search Algorithm for feature selection.

    Parameters:
    - HMS  : Harmony Memory Size
    - HMCR : Harmony Memory Considering Rate
    - PAR  : Pitch Adjusting Rate

    Steps:
    1. Initialize harmony memory with random binary masks.
    2. Evaluate all harmonies.
    3. Improvise a new harmony using HMCR, PAR, and random selection.
    4. Replace the worst harmony if the new harmony is better.
    5. Repeat for iterations.
    """

    num_features = X_train_data.shape[1]

    harmony_memory = np.random.randint(
        0,
        2,
        size=(harmony_memory_size, num_features)
    )

    for i in range(harmony_memory_size):
        harmony_memory[i] = ensure_non_empty_mask(harmony_memory[i])

    fitness_scores = np.array([
        evaluate_harmony(
            harmony,
            X_train_data,
            X_test_data,
            y_train_data,
            y_test_data
        )
        for harmony in harmony_memory
    ])

    best_index = np.argmax(fitness_scores)
    best_harmony = harmony_memory[best_index].copy()
    best_fitness = fitness_scores[best_index]

    fitness_history = []

    for iteration in range(iterations):

        new_harmony = np.zeros(num_features, dtype=int)

        for feature_idx in range(num_features):

            if np.random.rand() < harmony_memory_considering_rate:
                # Memory consideration
                selected_harmony_idx = np.random.randint(0, harmony_memory_size)
                new_harmony[feature_idx] = harmony_memory[selected_harmony_idx, feature_idx]

                # Pitch adjustment: flip bit
                if np.random.rand() < pitch_adjusting_rate:
                    new_harmony[feature_idx] = 1 - new_harmony[feature_idx]

            else:
                # Random selection
                new_harmony[feature_idx] = np.random.randint(0, 2)

        # Additional random exploration
        random_mask = np.random.rand(num_features) < random_selection_rate
        new_harmony[random_mask] = np.random.randint(
            0,
            2,
            size=np.sum(random_mask)
        )

        new_harmony = ensure_non_empty_mask(new_harmony)

        new_fitness = evaluate_harmony(
            new_harmony,
            X_train_data,
            X_test_data,
            y_train_data,
            y_test_data
        )

        worst_index = np.argmin(fitness_scores)

        if new_fitness > fitness_scores[worst_index]:
            harmony_memory[worst_index] = new_harmony.copy()
            fitness_scores[worst_index] = new_fitness

        current_best_index = np.argmax(fitness_scores)

        if fitness_scores[current_best_index] > best_fitness:
            best_fitness = fitness_scores[current_best_index]
            best_harmony = harmony_memory[current_best_index].copy()

        fitness_history.append(float(best_fitness))

        print(
            f"Iteration {iteration + 1}/{iterations} | "
            f"Best Fitness: {best_fitness:.4f} | "
            f"Selected Features: {int(np.sum(best_harmony))}"
        )

    return best_harmony, best_fitness, fitness_history


best_mask, best_fitness, fitness_history = hsa_feature_selection(
    X_train_dense,
    X_test_dense,
    y_train,
    y_test,
    harmony_memory_size=20,
    iterations=20,
    harmony_memory_considering_rate=0.90,
    pitch_adjusting_rate=0.30,
    random_selection_rate=0.10
)

best_mask = ensure_non_empty_mask(best_mask)

X_train_hsa = X_train_dense[:, best_mask == 1]
X_test_hsa = X_test_dense[:, best_mask == 1]

selected_feature_count = int(np.sum(best_mask))
total_feature_count = int(len(best_mask))

joblib.dump(best_mask, FEATURE_MASK_PATH)

print("\nHSA Feature Selection Completed.")
print("Total Processed Features:", total_feature_count)
print("Selected Features:", selected_feature_count)
print("Best HSA Fitness:", best_fitness)
print("HSA Feature Mask saved:", FEATURE_MASK_PATH)

# ============================================================
# 10. Deep Learning Model using HSA Features
# ============================================================

print("\nTraining HSA Deep Learning Model...")

input_dim = X_train_hsa.shape[1]
num_classes = len(label_encoder.classes_)

dl_model = Sequential([
    Dense(128, activation="relu", input_shape=(input_dim,)),
    Dropout(0.3),
    Dense(64, activation="relu"),
    Dropout(0.2),
    Dense(32, activation="relu"),
    Dense(num_classes, activation="softmax")
])

dl_model.compile(
    optimizer="adam",
    loss="sparse_categorical_crossentropy",
    metrics=["accuracy"]
)

early_stop = EarlyStopping(
    monitor="val_loss",
    patience=10,
    restore_best_weights=True
)

history = dl_model.fit(
    X_train_hsa,
    y_train,
    validation_split=0.2,
    epochs=100,
    batch_size=16,
    callbacks=[early_stop],
    verbose=1
)

dl_loss, dl_accuracy = dl_model.evaluate(X_test_hsa, y_test, verbose=0)

dl_pred_prob = dl_model.predict(X_test_hsa)
dl_predictions = np.argmax(dl_pred_prob, axis=1)

dl_model.save(H5_MODEL_PATH)

print("HSA Deep Learning Accuracy:", dl_accuracy)

# ============================================================
# 11. Machine Learning Models using HSA Features
# ============================================================

print("\nTraining Machine Learning Models using HSA-selected features...")

models = {
    "HSA Random Forest": RandomForestClassifier(
        n_estimators=200,
        random_state=42,
        class_weight="balanced",
        n_jobs=-1
    ),
    "HSA Logistic Regression": LogisticRegression(
        max_iter=1000,
        random_state=42
    ),
    "HSA SVM": SVC(
        kernel="rbf",
        probability=True,
        random_state=42
    )
}

model_results = []

best_model_name = None
best_model = None
best_accuracy = 0
best_predictions = None

for model_name, model in models.items():
    print(f"Training {model_name}...")

    model.fit(X_train_hsa, y_train)
    predictions = model.predict(X_test_hsa)
    accuracy = accuracy_score(y_test, predictions)

    model_results.append({
        "Model": model_name,
        "Accuracy": float(accuracy)
    })

    if accuracy > best_accuracy:
        best_accuracy = accuracy
        best_model_name = model_name
        best_model = model
        best_predictions = predictions

model_results.append({
    "Model": "HSA Deep Learning",
    "Accuracy": float(dl_accuracy)
})

if dl_accuracy > best_accuracy:
    overall_best_model_name = "HSA Deep Learning"
    overall_best_accuracy = float(dl_accuracy)
    overall_best_predictions = dl_predictions
else:
    overall_best_model_name = best_model_name
    overall_best_accuracy = float(best_accuracy)
    overall_best_predictions = best_predictions

joblib.dump(best_model, PKL_MODEL_PATH)
joblib.dump(preprocessor, PREPROCESSOR_PATH)
joblib.dump(label_encoder, LABEL_ENCODER_PATH)

print("\nBest ML Model:", best_model_name)
print("Best ML Accuracy:", best_accuracy)
print("Overall Best Model:", overall_best_model_name)
print("Overall Best Accuracy:", overall_best_accuracy)

# ============================================================
# 12. Result CSV
# ============================================================

result_df = pd.DataFrame(model_results)
result_df = result_df.sort_values(by="Accuracy", ascending=False)

result_df.to_csv(RESULT_CSV_PATH, index=False)

print("Result CSV saved:", RESULT_CSV_PATH)

# ============================================================
# 13. Prediction CSV
# ============================================================

prediction_df = X_test.copy()
prediction_df["Actual_Category"] = label_encoder.inverse_transform(y_test)
prediction_df["HSA_Best_Model"] = overall_best_model_name
prediction_df["HSA_Predicted_Category"] = label_encoder.inverse_transform(overall_best_predictions)
prediction_df["HSA_Deep_Learning_Prediction"] = label_encoder.inverse_transform(dl_predictions)

if best_predictions is not None:
    prediction_df["HSA_Best_ML_Prediction"] = label_encoder.inverse_transform(best_predictions)

prediction_df.to_csv(PREDICTION_CSV_PATH, index=False)

print("Prediction CSV saved:", PREDICTION_CSV_PATH)

# ============================================================
# 14. Accuracy Graph
# ============================================================

plt.figure(figsize=(10, 6))
sns.barplot(
    data=result_df,
    x="Model",
    y="Accuracy"
)
plt.title("HSA Model Accuracy Comparison")
plt.xlabel("Model")
plt.ylabel("Accuracy")
plt.xticks(rotation=30)
plt.ylim(0, 1)
plt.tight_layout()
plt.savefig(ACCURACY_GRAPH_PATH, dpi=300)
plt.show()

print("Accuracy graph saved:", ACCURACY_GRAPH_PATH)

# ============================================================
# 15. Heatmap
# ============================================================

cm = confusion_matrix(y_test, overall_best_predictions)

plt.figure(figsize=(8, 6))
sns.heatmap(
    cm,
    annot=True,
    fmt="d",
    cmap="Blues",
    xticklabels=label_encoder.classes_,
    yticklabels=label_encoder.classes_
)
plt.title(f"HSA Confusion Matrix Heatmap - {overall_best_model_name}")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.tight_layout()
plt.savefig(HEATMAP_PATH, dpi=300)
plt.show()

print("Heatmap saved:", HEATMAP_PATH)

# ============================================================
# 16. Comparison Graph
# ============================================================

plt.figure(figsize=(10, 6))
plt.plot(
    result_df["Model"],
    result_df["Accuracy"],
    marker="o",
    linewidth=2
)
plt.title("HSA Model Comparison Graph")
plt.xlabel("Model")
plt.ylabel("Accuracy")
plt.xticks(rotation=30)
plt.ylim(0, 1)
plt.grid(True)
plt.tight_layout()
plt.savefig(COMPARISON_GRAPH_PATH, dpi=300)
plt.show()

print("Comparison graph saved:", COMPARISON_GRAPH_PATH)

# ============================================================
# 17. Result Graph
# ============================================================

category_counts = df["performance_category"].value_counts()

plt.figure(figsize=(8, 6))
sns.barplot(
    x=category_counts.index,
    y=category_counts.values
)
plt.title("HSA Result Graph - Actual Performance Category Distribution")
plt.xlabel("Performance Category")
plt.ylabel("Number of Records")
plt.tight_layout()
plt.savefig(RESULT_GRAPH_PATH, dpi=300)
plt.show()

print("Result graph saved:", RESULT_GRAPH_PATH)

# ============================================================
# 18. Prediction Graph
# ============================================================

prediction_counts = prediction_df["HSA_Predicted_Category"].value_counts()

plt.figure(figsize=(8, 6))
sns.barplot(
    x=prediction_counts.index,
    y=prediction_counts.values
)
plt.title(f"HSA Prediction Distribution - {overall_best_model_name}")
plt.xlabel("Predicted Category")
plt.ylabel("Number of Predictions")
plt.tight_layout()
plt.savefig(PREDICTION_GRAPH_PATH, dpi=300)
plt.show()

print("Prediction graph saved:", PREDICTION_GRAPH_PATH)

# ============================================================
# 19. HSA Optimization Graph
# ============================================================

plt.figure(figsize=(10, 6))
plt.plot(
    range(1, len(fitness_history) + 1),
    fitness_history,
    marker="o",
    linewidth=2
)
plt.title("HSA Optimization Progress Graph")
plt.xlabel("Iteration")
plt.ylabel("Best Fitness")
plt.grid(True)
plt.tight_layout()
plt.savefig(OPTIMIZATION_GRAPH_PATH, dpi=300)
plt.show()

print("Optimization graph saved:", OPTIMIZATION_GRAPH_PATH)

# ============================================================
# 20. JSON Results
# ============================================================

best_report = classification_report(
    y_test,
    overall_best_predictions,
    target_names=label_encoder.classes_,
    output_dict=True,
    zero_division=0
)

dl_report = classification_report(
    y_test,
    dl_predictions,
    target_names=label_encoder.classes_,
    output_dict=True,
    zero_division=0
)

results_json = {
    "project_name": "Gram Panchayat Development Plan Performance Prediction System",
    "author": "Sagnik Patra",
    "algorithm": "Harmony Search Algorithm",
    "prefix": PREFIX,
    "dataset_path": DATASET_PATH,
    "output_directory": OUTPUT_DIR,
    "dataset_shape": {
        "rows": int(df.shape[0]),
        "columns": int(df.shape[1])
    },
    "target_column": target_column,
    "target_classes": list(label_encoder.classes_),
    "original_features": feature_columns,
    "hsa_feature_selection": {
        "total_processed_features": total_feature_count,
        "selected_features": selected_feature_count,
        "best_fitness": float(best_fitness),
        "fitness_history": [float(x) for x in fitness_history],
        "feature_mask_path": FEATURE_MASK_PATH,
        "harmony_memory_size": 20,
        "iterations": 20,
        "harmony_memory_considering_rate": 0.90,
        "pitch_adjusting_rate": 0.30,
        "random_selection_rate": 0.10
    },
    "best_model": {
        "model_name": overall_best_model_name,
        "accuracy": float(overall_best_accuracy),
        "classification_report": best_report,
        "confusion_matrix": cm.tolist()
    },
    "deep_learning_model": {
        "accuracy": float(dl_accuracy),
        "loss": float(dl_loss),
        "classification_report": dl_report
    },
    "all_model_results": result_df.to_dict(orient="records"),
    "saved_files": {
        "h5_model": H5_MODEL_PATH,
        "pkl_model": PKL_MODEL_PATH,
        "preprocessor": PREPROCESSOR_PATH,
        "label_encoder": LABEL_ENCODER_PATH,
        "feature_mask": FEATURE_MASK_PATH,
        "yaml_config": YAML_PATH,
        "json_results": JSON_PATH,
        "result_csv": RESULT_CSV_PATH,
        "prediction_csv": PREDICTION_CSV_PATH,
        "accuracy_graph": ACCURACY_GRAPH_PATH,
        "heatmap": HEATMAP_PATH,
        "comparison_graph": COMPARISON_GRAPH_PATH,
        "result_graph": RESULT_GRAPH_PATH,
        "prediction_graph": PREDICTION_GRAPH_PATH,
        "optimization_graph": OPTIMIZATION_GRAPH_PATH
    }
}

with open(JSON_PATH, "w") as f:
    json.dump(results_json, f, indent=4)

print("JSON results saved:", JSON_PATH)

# ============================================================
# 21. YAML Config
# ============================================================

config_yaml = {
    "project": {
        "name": "Gram Panchayat Development Plan Performance Prediction System",
        "author": "Sagnik Patra",
        "description": "Predicts GPDP performance category using HSA-based feature selection, ML, and deep learning."
    },
    "algorithm": {
        "name": "Harmony Search Algorithm",
        "short_name": "HSA",
        "type": "Bio-Inspired Optimization",
        "purpose": "Feature Selection",
        "harmony_memory_size": 20,
        "iterations": 20,
        "harmony_memory_considering_rate": 0.90,
        "pitch_adjusting_rate": 0.30,
        "random_selection_rate": 0.10,
        "selected_features": selected_feature_count,
        "total_processed_features": total_feature_count,
        "best_fitness": float(best_fitness)
    },
    "dataset": {
        "path": DATASET_PATH,
        "target_column": target_column,
        "original_features": feature_columns,
        "categorical_features": categorical_features,
        "numeric_features": numeric_features
    },
    "models": {
        "machine_learning_models": list(models.keys()),
        "best_ml_model": best_model_name,
        "best_ml_accuracy": float(best_accuracy),
        "overall_best_model": overall_best_model_name,
        "overall_best_accuracy": float(overall_best_accuracy),
        "deep_learning": {
            "type": "Keras Sequential Neural Network",
            "accuracy": float(dl_accuracy),
            "loss": float(dl_loss),
            "optimizer": "adam",
            "loss_function": "sparse_categorical_crossentropy",
            "epochs": 100,
            "batch_size": 16
        }
    },
    "outputs": {
        "h5_model": H5_MODEL_PATH,
        "pkl_model": PKL_MODEL_PATH,
        "preprocessor": PREPROCESSOR_PATH,
        "label_encoder": LABEL_ENCODER_PATH,
        "feature_mask": FEATURE_MASK_PATH,
        "json_results": JSON_PATH,
        "result_csv": RESULT_CSV_PATH,
        "prediction_csv": PREDICTION_CSV_PATH,
        "accuracy_graph": ACCURACY_GRAPH_PATH,
        "heatmap": HEATMAP_PATH,
        "comparison_graph": COMPARISON_GRAPH_PATH,
        "result_graph": RESULT_GRAPH_PATH,
        "prediction_graph": PREDICTION_GRAPH_PATH,
        "optimization_graph": OPTIMIZATION_GRAPH_PATH
    }
}

with open(YAML_PATH, "w") as f:
    yaml.dump(config_yaml, f, default_flow_style=False)

print("YAML config saved:", YAML_PATH)

# ============================================================
# 22. Final Summary
# ============================================================

print("\n============================================================")
print("HSA PROJECT EXECUTION COMPLETED SUCCESSFULLY")
print("============================================================")
print("Saved Files:")
print("H5 Model:", H5_MODEL_PATH)
print("PKL Model:", PKL_MODEL_PATH)
print("Preprocessor:", PREPROCESSOR_PATH)
print("Label Encoder:", LABEL_ENCODER_PATH)
print("HSA Feature Mask:", FEATURE_MASK_PATH)
print("YAML Config:", YAML_PATH)
print("JSON Results:", JSON_PATH)
print("Result CSV:", RESULT_CSV_PATH)
print("Prediction CSV:", PREDICTION_CSV_PATH)
print("Accuracy Graph:", ACCURACY_GRAPH_PATH)
print("Heatmap:", HEATMAP_PATH)
print("Comparison Graph:", COMPARISON_GRAPH_PATH)
print("Result Graph:", RESULT_GRAPH_PATH)
print("Prediction Graph:", PREDICTION_GRAPH_PATH)
print("Optimization Graph:", OPTIMIZATION_GRAPH_PATH)
print("============================================================")
```
