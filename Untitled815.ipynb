```python
import os
import json
import yaml
import joblib
import warnings
import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder, StandardScaler, LabelEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.ensemble import RandomForestClassifier

import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.callbacks import EarlyStopping

warnings.filterwarnings("ignore")

# ============================================================
# 1. Paths
# ============================================================

DATASET_PATH = r"C:\Users\NXTWAVE\Downloads\Gram Panchayat Development Plan Performance Prediction System\37ca30d3-ea8c-48a5-a5dc-50ec5f46edd5.csv"

OUTPUT_DIR = r"C:\Users\NXTWAVE\Downloads\Gram Panchayat Development Plan Performance Prediction System"

os.makedirs(OUTPUT_DIR, exist_ok=True)

H5_MODEL_PATH = os.path.join(OUTPUT_DIR, "gpdp_performance_model.h5")
PKL_MODEL_PATH = os.path.join(OUTPUT_DIR, "gpdp_random_forest_model.pkl")
PREPROCESSOR_PATH = os.path.join(OUTPUT_DIR, "gpdp_preprocessor.pkl")
LABEL_ENCODER_PATH = os.path.join(OUTPUT_DIR, "gpdp_label_encoder.pkl")
YAML_CONFIG_PATH = os.path.join(OUTPUT_DIR, "gpdp_config.yaml")
JSON_RESULT_PATH = os.path.join(OUTPUT_DIR, "gpdp_results.json")
PREDICTION_CSV_PATH = os.path.join(OUTPUT_DIR, "gpdp_predictions.csv")
RESULT_CSV_PATH = os.path.join(OUTPUT_DIR, "gpdp_result_summary.csv")

# ============================================================
# 2. Load Dataset
# ============================================================

print("Loading dataset...")

df = pd.read_csv(DATASET_PATH)

print("Dataset loaded successfully.")
print("Shape:", df.shape)
print("Columns:", list(df.columns))

# ============================================================
# 3. Basic Cleaning
# ============================================================

df.columns = df.columns.str.strip()

required_columns = [
    "stateCode",
    "stateName",
    "districtCode",
    "districtName",
    "finYear",
    "totalGP",
    "sabhaHeld",
    "gpdpApproved",
    "gpdpUpload",
    "date updated"
]

missing_columns = [col for col in required_columns if col not in df.columns]

if missing_columns:
    raise ValueError(f"Missing columns in dataset: {missing_columns}")

# Convert date column
df["date updated"] = pd.to_datetime(df["date updated"], errors="coerce")

# Fill missing dates with mode or default date
if df["date updated"].isnull().sum() > 0:
    df["date updated"] = df["date updated"].fillna(df["date updated"].mode()[0])

# Extract date features
df["updated_year"] = df["date updated"].dt.year
df["updated_month"] = df["date updated"].dt.month
df["updated_day"] = df["date updated"].dt.day

# Avoid division by zero
df["totalGP"] = df["totalGP"].replace(0, np.nan)

# Fill missing numeric values
numeric_cols = ["totalGP", "sabhaHeld", "gpdpApproved", "gpdpUpload"]
for col in numeric_cols:
    df[col] = df[col].fillna(df[col].median())

# Fill missing categorical values
categorical_cols = ["stateName", "districtName", "finYear"]
for col in categorical_cols:
    df[col] = df[col].fillna("Unknown")

# ============================================================
# 4. Feature Engineering
# ============================================================

print("Creating performance score and target category...")

df["sabha_ratio"] = df["sabhaHeld"] / df["totalGP"]
df["approval_ratio"] = df["gpdpApproved"] / df["totalGP"]
df["upload_ratio"] = df["gpdpUpload"] / df["totalGP"]

df["performance_score"] = (
    df["sabha_ratio"] +
    df["approval_ratio"] +
    df["upload_ratio"]
) / 3

# Clip score between 0 and 1 for safety
df["performance_score"] = df["performance_score"].clip(0, 1)

# Create target category
def classify_performance(score):
    if score < 0.50:
        return "Low"
    elif score < 0.80:
        return "Medium"
    else:
        return "High"

df["performance_category"] = df["performance_score"].apply(classify_performance)

print("Target distribution:")
print(df["performance_category"].value_counts())

# ============================================================
# 5. Features and Target
# ============================================================

feature_columns = [
    "stateCode",
    "stateName",
    "districtCode",
    "districtName",
    "finYear",
    "totalGP",
    "sabhaHeld",
    "gpdpApproved",
    "gpdpUpload",
    "updated_year",
    "updated_month",
    "updated_day",
    "sabha_ratio",
    "approval_ratio",
    "upload_ratio"
]

target_column = "performance_category"

X = df[feature_columns]
y = df[target_column]

# Encode target labels
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)

print("Label classes:", list(label_encoder.classes_))

# ============================================================
# 6. Preprocessing
# ============================================================

categorical_features = ["stateName", "districtName", "finYear"]

numeric_features = [
    "stateCode",
    "districtCode",
    "totalGP",
    "sabhaHeld",
    "gpdpApproved",
    "gpdpUpload",
    "updated_year",
    "updated_month",
    "updated_day",
    "sabha_ratio",
    "approval_ratio",
    "upload_ratio"
]

preprocessor = ColumnTransformer(
    transformers=[
        ("num", StandardScaler(), numeric_features),
        ("cat", OneHotEncoder(handle_unknown="ignore"), categorical_features)
    ]
)

# ============================================================
# 7. Train-Test Split
# ============================================================

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y_encoded,
    test_size=0.2,
    random_state=42,
    stratify=y_encoded
)

# Fit preprocessor
X_train_processed = preprocessor.fit_transform(X_train)
X_test_processed = preprocessor.transform(X_test)

# Convert sparse matrix to dense for TensorFlow
X_train_dense = X_train_processed.toarray() if hasattr(X_train_processed, "toarray") else X_train_processed
X_test_dense = X_test_processed.toarray() if hasattr(X_test_processed, "toarray") else X_test_processed

num_classes = len(label_encoder.classes_)
input_dim = X_train_dense.shape[1]

print("Processed input dimension:", input_dim)
print("Number of classes:", num_classes)

# ============================================================
# 8. Train Deep Learning Model and Save H5
# ============================================================

print("Training deep learning model...")

model = Sequential([
    Dense(128, activation="relu", input_shape=(input_dim,)),
    Dropout(0.3),
    Dense(64, activation="relu"),
    Dropout(0.2),
    Dense(32, activation="relu"),
    Dense(num_classes, activation="softmax")
])

model.compile(
    optimizer="adam",
    loss="sparse_categorical_crossentropy",
    metrics=["accuracy"]
)

early_stop = EarlyStopping(
    monitor="val_loss",
    patience=10,
    restore_best_weights=True
)

history = model.fit(
    X_train_dense,
    y_train,
    validation_split=0.2,
    epochs=100,
    batch_size=16,
    callbacks=[early_stop],
    verbose=1
)

dl_loss, dl_accuracy = model.evaluate(X_test_dense, y_test, verbose=0)

print("Deep Learning Test Accuracy:", dl_accuracy)

model.save(H5_MODEL_PATH)
print("H5 model saved at:", H5_MODEL_PATH)

# ============================================================
# 9. Train Random Forest Model and Save PKL
# ============================================================

print("Training Random Forest model...")

rf_model = RandomForestClassifier(
    n_estimators=200,
    random_state=42,
    max_depth=None,
    class_weight="balanced"
)

rf_model.fit(X_train_processed, y_train)

rf_predictions = rf_model.predict(X_test_processed)
rf_accuracy = accuracy_score(y_test, rf_predictions)

print("Random Forest Accuracy:", rf_accuracy)

joblib.dump(rf_model, PKL_MODEL_PATH)
joblib.dump(preprocessor, PREPROCESSOR_PATH)
joblib.dump(label_encoder, LABEL_ENCODER_PATH)

print("PKL model saved at:", PKL_MODEL_PATH)
print("Preprocessor saved at:", PREPROCESSOR_PATH)
print("Label encoder saved at:", LABEL_ENCODER_PATH)

# ============================================================
# 10. Predictions
# ============================================================

print("Generating predictions...")

dl_pred_prob = model.predict(X_test_dense)
dl_predictions = np.argmax(dl_pred_prob, axis=1)

prediction_df = X_test.copy()
prediction_df["actual_category"] = label_encoder.inverse_transform(y_test)
prediction_df["dl_predicted_category"] = label_encoder.inverse_transform(dl_predictions)
prediction_df["rf_predicted_category"] = label_encoder.inverse_transform(rf_predictions)

prediction_df.to_csv(PREDICTION_CSV_PATH, index=False)

print("Prediction CSV saved at:", PREDICTION_CSV_PATH)

# ============================================================
# 11. Reports
# ============================================================

dl_report = classification_report(
    y_test,
    dl_predictions,
    target_names=label_encoder.classes_,
    output_dict=True
)

rf_report = classification_report(
    y_test,
    rf_predictions,
    target_names=label_encoder.classes_,
    output_dict=True
)

dl_confusion = confusion_matrix(y_test, dl_predictions).tolist()
rf_confusion = confusion_matrix(y_test, rf_predictions).tolist()

results = {
    "project_name": "Gram Panchayat Development Plan Performance Prediction System",
    "author": "Sagnik Patra",
    "dataset_path": DATASET_PATH,
    "output_directory": OUTPUT_DIR,
    "total_rows": int(df.shape[0]),
    "total_columns": int(df.shape[1]),
    "features_used": feature_columns,
    "target_column": target_column,
    "target_classes": list(label_encoder.classes_),
    "deep_learning_model": {
        "model_type": "Keras Sequential Neural Network",
        "h5_model_path": H5_MODEL_PATH,
        "test_accuracy": float(dl_accuracy),
        "test_loss": float(dl_loss),
        "classification_report": dl_report,
        "confusion_matrix": dl_confusion
    },
    "random_forest_model": {
        "model_type": "RandomForestClassifier",
        "pkl_model_path": PKL_MODEL_PATH,
        "test_accuracy": float(rf_accuracy),
        "classification_report": rf_report,
        "confusion_matrix": rf_confusion
    },
    "saved_files": {
        "h5_model": H5_MODEL_PATH,
        "pkl_model": PKL_MODEL_PATH,
        "preprocessor": PREPROCESSOR_PATH,
        "label_encoder": LABEL_ENCODER_PATH,
        "yaml_config": YAML_CONFIG_PATH,
        "json_result": JSON_RESULT_PATH,
        "prediction_csv": PREDICTION_CSV_PATH,
        "result_csv": RESULT_CSV_PATH
    }
}

# Save JSON result
with open(JSON_RESULT_PATH, "w") as f:
    json.dump(results, f, indent=4)

print("JSON result saved at:", JSON_RESULT_PATH)

# ============================================================
# 12. YAML Config
# ============================================================

config = {
    "project": {
        "name": "Gram Panchayat Development Plan Performance Prediction System",
        "author": "Sagnik Patra",
        "description": "Predicts Gram Panchayat Development Plan performance category using ML and deep learning."
    },
    "paths": {
        "dataset_path": DATASET_PATH,
        "output_directory": OUTPUT_DIR,
        "h5_model_path": H5_MODEL_PATH,
        "pkl_model_path": PKL_MODEL_PATH,
        "preprocessor_path": PREPROCESSOR_PATH,
        "label_encoder_path": LABEL_ENCODER_PATH,
        "json_result_path": JSON_RESULT_PATH,
        "prediction_csv_path": PREDICTION_CSV_PATH,
        "result_csv_path": RESULT_CSV_PATH
    },
    "data": {
        "features": feature_columns,
        "target": target_column,
        "categorical_features": categorical_features,
        "numeric_features": numeric_features
    },
    "model": {
        "deep_learning": {
            "type": "Keras Sequential",
            "layers": [
                "Dense 128 ReLU",
                "Dropout 0.3",
                "Dense 64 ReLU",
                "Dropout 0.2",
                "Dense 32 ReLU",
                "Dense Softmax"
            ],
            "optimizer": "adam",
            "loss": "sparse_categorical_crossentropy",
            "epochs": 100,
            "batch_size": 16
        },
        "machine_learning": {
            "type": "RandomForestClassifier",
            "n_estimators": 200,
            "random_state": 42,
            "class_weight": "balanced"
        }
    },
    "performance": {
        "deep_learning_accuracy": float(dl_accuracy),
        "random_forest_accuracy": float(rf_accuracy)
    }
}

with open(YAML_CONFIG_PATH, "w") as f:
    yaml.dump(config, f, default_flow_style=False)

print("YAML config saved at:", YAML_CONFIG_PATH)

# ============================================================
# 13. Result Summary CSV
# ============================================================

summary_df = pd.DataFrame({
    "Model": ["Deep Learning Neural Network", "Random Forest"],
    "Accuracy": [float(dl_accuracy), float(rf_accuracy)],
    "Model File": [H5_MODEL_PATH, PKL_MODEL_PATH]
})

summary_df.to_csv(RESULT_CSV_PATH, index=False)

print("Result summary CSV saved at:", RESULT_CSV_PATH)

# ============================================================
# 14. Final Output
# ============================================================

print("\n============================================================")
print("PROJECT EXECUTION COMPLETED SUCCESSFULLY")
print("============================================================")
print("Saved files:")
print("H5 Model:", H5_MODEL_PATH)
print("PKL Model:", PKL_MODEL_PATH)
print("Preprocessor PKL:", PREPROCESSOR_PATH)
print("Label Encoder PKL:", LABEL_ENCODER_PATH)
print("YAML Config:", YAML_CONFIG_PATH)
print("JSON Results:", JSON_RESULT_PATH)
print("Prediction CSV:", PREDICTION_CSV_PATH)
print("Result CSV:", RESULT_CSV_PATH)
print("============================================================")
```
